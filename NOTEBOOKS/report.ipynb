## 1. Introduction 

### 1.1 Literature review of the compound (<500 words)

The amino acid L-lysine is considered as one of the nine essential proteins building blocks for livestocks and humans beings because it is a fundamental unit for all proteins in the body. As a consequence, L-lysine products has showed a huge uptick in the market for the last decade. Nowadays, L-lysine shows increasing applications in different fields, such as, supplementation in sustenance and animal feed in nutritionally correct amounts. Moreover, it is also used as a chemical ingredient in the cosmetics and pharmaceutical industries, while sold as a diet supplements in retail stores in different formulations [1, 2].

Industrially, L-lysine is biomanufactured via fermentation, completely replacing the chemical synthesis which was not competitive since 1978 [3]. Its final formulation is a stable, non-hygroscopic hydrochlorinated form of L-lysine-HCl, with a chemical formula of H2N(CH2)4CHNH2CO2H-HCl [4]. The mean global price of L-lysine for animal feed has been 1.5 - 2.0 USD/kg for the past decade. Annual global production of L-lysine is around one million tons and is expected to rise in the early future due to higher demand in meat-containing diets around the world, mostly in Asia and Latin America [5].

Furthermore, non-meat plant-based feedstuff is restricted in essential amino acids; as a result, supplementation is required to ensure a healthy growth of animals. L-lysine, and other essential amino acids, ensures a rapid and robust development of the animal; without it, animal development is limited even if excess nitrogen is available in their diets [5].

Inside our bodies, L-lysine participates in  recovering from sports injuries or surgeries, building muscle protein and calcium absorption, but also, in bodys production of hormones, enzymes, and antibodies.

The biosynthetic pathway of L-lysine in _C. glutamicum_, for instance, is depicted in figure 1. The synthsis can be summarized in two main routes which lead into the L-lysine production in the bacterium. Firstly, the aspartate biosynthetic pathway, with its begging in oxaloacetic acid (OAA). Secondly, the diaminopimelic acid (DAP) and -aminoadipic acid (AAA) pathways, where the DAP pathway includes the conversion of OAA to L-aspartate, while the AAA pathway involves the synthesis of L-lysine from L-aspartate [6]. 

In the present report, it has been decided to target either the DAPDC reaction, for its convergent location between pathways inducing amino acid accumulation in the cell, or the secreation of L-lysine, depending on the section, as a model objective reaction.

### 1.2 Literature review of the cell factory 


Microbial fermentation is the currently preferred method for the industrial production of L-lysine, based on the bacterium _Corynebacterium glutamicum_ (_C. glutamicum_) as the cell factory. Fermentation has completely replaced chemical synthesis due to its greater cost-effectiveness, sustainability, and scalability [2].

_C. glutamicum_ is a Gram-positive bacterium that naturally produces amino acids, and it has been extensively studied and engineered to enhance a large range of compounds production rates. Various strategies, such as genome-scale modeling, metabolic engineering, and gene knockouts, have been explored to optimize the biosynthesis pathways in _C. glutamicum_, aiming to develop high-yield strains capable of efficiently and specifically producing amino acids, for instance [2]. A variety of genetically engineered _E. coli_ and _C. glutamicum_ strains are available in the literature [7]. 

In this study, a _C. glutamicum_ Genome Scale Modeling (GEM) engineering approach will be used to potentially enhance its ability to synthesize L-lysine.

On the one hand, the bacterium presents a series of advantages as a cell factory, becoming an industry workhorse. Firstly, its independence from external glutamate supply and complex media components. Secondly, the bacterium shows high versatility as a cell factory for the production of useful bulk chemicals under both aerobic and anaerobic conditions. For instance, organic acids (isobutanol, ethanol, 3-methyl-1-butanol, 2,3-butanediol) and amino acids (L-lysine, L-alanine, L-glutamate, and L-valine) [8].

On the other hand, some disadvantages have been identified when using _C. glutamicum_ as an industrial cell factory. Firstly, the bacterium has limitations in the production of complex proteins [9]. Secondly, metabolic engineering of _C. glutamicum_ is still challenging [10]. Finally, some drawbacks exist associated with using the bacterium in L-threonine biosynthesis [11]. 

Industrial fermentation with _C. glutamicum_ is based on using glucose as a feedstock, being a native and prefered substrate for the bacteria. Glucose has been used extensively in biotechnology due to its place in the central metabolism, making it easy to integrate into pathways; its low price; and the ease of manufacturing from common agricultural products [2]. 

## 2. Problem definition (<300 words)

As said before, currently L-lysine production is based on glucose as a feedstock. Nonetheless, glucose poses environmental challenges due to its resource-intensive purification process from agricultural polysaccharides [12]. Moreover, its production competes with food manufacture. This approach has detrimental effects on nature, public health, and ethical considerations in the face of food shortages, specially upon climate change and the current global energy crisis [13, 14, 15, 16].

Consequently, research is shifting towards alternative feedstocks for _C. glutamicum_, with glycerol emerging as a more sustainable option than glucose. Glycerol (Rittmann et al., 2008), a byproduct of various industrial processes, offers an environmentally friendly and cost-effective alternative for L-lysine production [17].

Furthermore, to meet the L-lysine market needs, the industrial production of the amino acid must use robust _C. glutamicum_ strains which present efficient productivity and economic production costs. Even though that industrially currently used strains are showing promising productivities, the scientific community is providing intense efforts on reasearching and developing better strains on a constant-basis [6]. For such a purpose, the use of good and corraborated GSM technique can be a shortcut on the process of testing new strains, reducing the required time and number of experiments in the laboratory. 

This report aims to review already existing GSM model for the bacteria _C. glutamicum_ focusing on the production of L-lysine, assesing if their feasibility to be used to engineer and simulate potential improvements for a hypotehtical biomanufacturing of this amino acid. For such a purpose, **the document has been divided in two main tasks**:

1- Evaluate **alternative feedstocks to glucose** that can yield better results, specifically glycerol. 

2- **Metabolic engineering of the L-lysine pahtway** by assessing the selected GSM, and reproducing genetic modifications aimed to improving L-lysine production from _C. glutamicum_. 

**Are the studied GSMs for _C. glutamicum_ optimal to be considered as a platform to further reasearch the improvment of L-lysine industrial production?**

## 3. Selection and assessment of existing GSM

Two GSM models for the bacteria _Corynebacterium glutamicum_ ATCC13032 were assessed in the present report to potentially propose improvements in L-lysine production. 

Zhang et al., 2017 [19] constructed a GSM model for the bacteria _Corynebacterium glutamicum_ ATCC 13032 (iCW773) motivated by the powerful industrial organism platform the bacteria represents, as described above. This model, known as iCW773, contains 773 genes, 950 metabolitesa and 1207 reactions. Pearsons correlation coefficient between the iCW773-simulated and experimental fluxes was 0.99, suggesting that iCW773 provided an accurate intracellular flux distribution of the wild-type strain growing on glucose.

Feierabend et al., 2021 [18] developed a GEM model for the _Corynebacterium glutamicum_(iCGB21FR), specifically for the strain ATCC 13032 which is extensively used in the biotechnology sector to produce amino acids in the large scale. It offers several avantages such as an apparent robustness - becoming a suitable platform host for severl compunds, for instance L-lysine. The research team updated and unified the Genome-Scale Metabolic model (GEM) of _C. glutamicum_ ATCC 1302, employing high-quality, state-of-the-art modeling techniques and advanced reconstruction algorithms. The resultant GEM iCGB21FR presents 1042 metabolites, 1539 reactions and 805 genes. Its phenotypic predictions resulted highly consistent with the laboratory data obtained with different media and both aerobic and anaerobic growth conditions.

After an extensive literature research, these two GSM were found to be considred as the most uptaded for the strain  _Corynebacterium glutamicum_ ATCC 1303. 

As a first comparison and characteritzation of the selected GSM, a **memote** for each one is performed and the results depicted in figure 2. The table suggested that iCGB21FR is a more representative model of _C. glutamicum_ ATCC 1302 showing a score of 70%, whereas iCW773 showed only 23%. As a result, both models are detailed and used when simulating fermentations with alternatives feedstock; but only iCGB21FR is used for the metabolic engineering part.

### 4.1 Characteritzation of _Corynebacterium glutamicum_ ATCC 13032 GSM 

_Corynebacterium glutamicum_ can produce L-Lysine under a specific growth conditions. Here, a GSM-based approach is used to study the bioprocess. Firstly, each model for _C. glutamicum_ ATCC 13032, iCW773 (2017) and iCGB21FR (2021), are characterized in detail.

To achieve it, the MEMOTE tool was used, generating information that detailed the quality of each of the models. As presented in figure 2, the MEMOTE analysis of iCW773 showed a low quality of the model, and an inconsistency in the objective function, being unknown what had to be optimized. 

On the other hand, the model iCGB21FR presented a better quality compared to the first one. When deciding which model to use in each activity, we took into account the errors presented in each of the models and their quality. 

In order to study the optimal growth of _Corynebacterium glutamicum_ ATCC 13032 for both GSMs when producing L-lysine, a Phenotypic Phase Plane (PPP) analysis is used to provide a view of the relation between two exchange reactions. 

The PPP analysis was run under aerobic and anaerobic conditions. This is because _C. glutamicum_ being is a facultative aerobic gram-positive bacterium that is able to growth in the presence of oxygen (aerobic) as well as in the absence of oxygen (anaerobic). 

#### Aerobic conditions
Beginning with iCW773, the PPP analysis are plotted (figure 3) for the biomass as a function of lysine exchange reaction, exchange reaction of lysine as a function of glucose uptake, as well as lysine production as a function of oxygen exchange and biomass production as a function of DAPDC, all for both oxygen conditions. 

 
 

Figure 3: Phenotypic phase planes under aerobic conditions - iCW773

In the model iCW773 it is observed that there is a correlation between biomass production and lysine production. The more carbon goes to lysine production, the less goes to biomass. While consuming more glucose, the production of lysine increases. According to the third figure there is low lysine production when there is little oxygen. It produces a high flux in a range of oxygen, before falling again at higher level of oxygen. Finally, in the fourth figure the biomass starts to incresase while the production of DAPDC it's low. Then the carbon goes to DAPDC reaction, decreasing the biomass.

The model iCGB21FR did not presented a production of lysine compared to different biomass concentrations and oxygen ranges (figure 4). The same behavior was evident with the glucose exchange reaction, in which this carbon souce increases immeediately.In the fourth figure there is a correlation between biomass production and DAPDC reaction. The more carbon goes to DAPDC reaction, the more carbon invested to biomass.

 

 

Figure 4: Phenotypic phase planes under aerobic conditions - iCGB21FR

According to these results of the PPP and the modeling analysis of iCGB21FR, it is noted that this model does not incorporate a Secretion system to export lysine outside of the cell once it is accumulated as a result of DAPDC. However, iCGB21FR presents a better quality and details compared to the other model (figure 2). As a consequence, a L-lysine secretion system is added to the model in section 4.4. 

#### Anaerobic conditions

Lysine production is an aerobic process, thus, the computational analysis under anaerobic conditions yielded negative results in both models.

### 4.2 Optimitzation of the carbon source

As outlined earlier, the current use of glucose as the feedstock for L-lysine fermentation faces sustainability and availability challenges. To address this, various alternative carbon sources have been proposed for use with _C. glutamicum_ as the cell factory.

The two selected GSM for the cell factory, either iCW773 or iCGB21F, were used to predict alternative carbon options for the fermentation. The DADPC reaction is set as the model objective for the following analysis, targeting the accumulation of amino acid inside the cell. 

Initially, individual carbon sources were tested in the abscence of glucose from the media. Subsequently, the maximal growth rates and L-Lysine production for each substrate were documented and ranked in a table. 

Starting with the GSM iCW773, it was observed that all potential carbon alternatives to glucose completely inhibited growth and, subsequently, amino acid synthesis. Therefore, no viable feedstock alternatives could be suggested for the fermentation based on the iCW773 model. At this point, the model was considered not optimal for this kind of investigation. 

Secondly, the same analysis was carried on with the iCGB21FR model. As a result, on several carbon substrates, growth and production rates, were found greater than 0, as ilustrated in figure 5. 



Figure 5: Plot showing the maximal growth rate and maximal L-lysine production rate for all carbon exchange reactions in the model, when glucose was not available - iCGB21FR.

When comparing the maximal growth and L-lysine production rates with alternative feedstocks to the basal values on glucose in iCGB21FR, it was observed that **only compounds made of glucose subunits**, such as amylose, glycogen, and starch, led to a significant improvement in the rates. Based on these results from iCGB21FR, it was suggested that glucose is the preferred and possibly the best feedstock for the bacteria when producing L-lysine. Therefore, locally in this model, providing more glucose through its polymers improved the yields.



Figure 6: Growth ranked table of the top 15 alternative feedstock, ahowing also the predicted L-lysine production rate. The basal growth and production rates when using glucose were 0.571 and 0.189 mmol/gDW*h*, respectively - iCGB21FR.

Among all possible alternatives, glycerol stands out as a promising sustaniable feedstock [17, 20]. Although, **glycerol was not predicted as a promising alternative** in the ranked table obtained above, it was decided to further investigated this individual carbon source to corraborate its non-viability. Following discussion is based only in the GSM iCGB21FR. 

_C. glutamicum_ can natively uptake and use glycerol as feedstock via three reactions before it is incorporate as dihydroxyacetone phosphate into the glycolysis pathway. However, it is known that the growth rate and efficiency of glycerol utilization can be improved by introducing additional genes or pathways from other organisms, such as _E. coli_ [20]. After investigating if iCGB21FR already incorporated the utilitzation pathway, it was concluded that the model already incorporated all the reactions but the glycerol was exchanged from the media in form of Glycerol 3-phosphate, skiping the initial phophorilation after the uptake. The uptake reaction presents a rate 1:1 between extracellular glycerol and Glycerol 3-phosphate, so the last compound is used to simulate the glycerol feedstock.

After this validation, the original medium was modified to contain glycerol instead of glucose. The maximal theoretical productivity of lysine with the new medium was 0.136 mmol/gDW*h* whereas the basal value when using glucose was 0.189 mmol/gDW*h*. These results pointed to the same direction; iCGB21FR predicted that the feedstock might not be easy to be substituted in a native cell factory, at least without trading-off other yields of the process.  

Finally, a Phenotypic phase plane anlysis on the production of L-lysine and biomass generation on glycerol was performed (figure 4). Comparing its tendency with the one obtained from growth on glucose (figure 7), it suggested that the produciton of L-lysine on glycerol is still coupled to the biomass generation, but in lower rate.

### 4.3 Finding the name of genes in the model iCGB21FR 

Our model implemented **a non standard naming format** which prevented us from easily finding our target genes for the metabolic engineering part. The genes were named in this format:
 **"lcl_{locus}_prot_{protein_id}"**
 
An exemple name of the genes from the model would look like this: lcl_NC_006958_1_prot_WP_003859586_1_2138

In which: 
- **"NC_006958"** is the **locus ID** that represent the complete sequence of _Corynebacterium glutamicum_ ATCC 13032, and
- **WP_XXXXXXXXX** represent the **protein ID** of its gene.

This nomenclature made slow and circuitous any operation that involded genes.

For this reason, a pipeline was devolleped in order to allow us to **match the conventianal names of genes found in literature with the names used in the model**.

First, all the genes found on **RefSeq database** from _Corynebacterium glutamicum_ ATCC 13032 were downloaded and used to search across the annonatated genes that have a gene name.

For the missing genes that did not have **a direct match between the protein ID and the target gene**, a search in other _Corynebacterium glutamicum_ strains** was performed, hoping that the target gene would have been at least **annotated** in one of those strains. 

Once the target gene is found, its corresponding AA sequence is taken and searched against all the genes from _Corynebacterium glutamicum_ ATCC 13032 in order to find the protein ID.


Checking the KEGG database resulted in finding 10 homologous _Corynebacterium glutamicum_ strains.

**This technique turned out to be sucessfull in finding all the genes of our interet.**

### 4.4 Adding Lysine export reaction

As seen during the phenotypique phase plane analysis, **the model iCGB21FR cannot secret lysine**. Thus, adding this reaction was necessary.

In the literature, it has been found that multiple exporters exists(e.g YbjE), **but no information about the stochiometry of the export reaction were available**.

## 5. Discussion

The objective of this report was to assess the viability of using two different model for the bacteria _C. glutamicum_ ATCC 13032, specifically iCW773 (from 2017) and iCGB21FR (from 2021), as an approach to engineer improvement to the current L-lysine fermentation based biomanufacturing. 

This fermentation has two main problems identified. Firstly, its feedstock is glucose, while it is considered as not sustainable carbon source. Secondly, the annual global production of L-lysine is around one million tons and is expected to rise. As a result, the development of more productive and robust cell factories is really necessary. 

On the one hand, alternative feedstock were used to simulate the L-lysine fermentation, switching the orginal glucose. iCW773 predicted no other carbon soruce than glucose whereas iCGB21FR predicted some alternative. However, the other feedstock resulted in growth and L-lysine rates lower than the original ones with glucose. As a conclusion, it suggested that changing the feedstock on L-lysine production may be quite difficult to simulate. 

On the other hand, the model iCGB21FR was used to reproduce some succesful metabolic changes described in the literature. It was expected to get similar results on the L-lysine productivity after applying the sames changes. However, after mapping all the genes in the model to facilitate its identification and mimicks all the genetic changes, no better or worse L-lysine synthesis rates were observed. Parhaps, the inviability of reproducing the reference article is due to the bounds of the reactions, here they were setted randomly, lacking from _in vivo_ reference. 

## 6. Conclusion 

In this project, a GSM approach has been assessed for its feasibility as a technique to simulate and research improvement to the current L-lyisine fermentation process. The bioprocess is currently based on _Corynebacterium glutamicum_ bacterium and glucose as feedstock. However, the glucose feedstock, which is considered not sustainable, and the need to develop more productive and robust cell factories to meet the market needs, have been the issues addressed here. 

Two GSM for the _C. glutamicum_ ATCC 13032, iCW773 (2017) and iCGB21FR (2021), have been used to simulate and predict alternative feedstock for the fermentation and reproduce some genetic modification on the L-lysine pathways carried out in the literature. As a result, the feasibility of using these GSM as a platform to further investigate the L-lysine production has been interrogated. 

Overall, the results of the two analysis look promising, even though some problemes arised. Together with further research and improvement of the model, GSM approach represent a useful tool to design better L-lysine manufacturing processes.
