**Two .html file reports** were generated


```python
from ipywidgets import interact, widgets, Layout
from IPython.display import display, HTML

def viewHTMLFile(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()
        display(HTML(html_content))
        
def selectFile(option):
    if option == 'iCBG21FR.html':
        viewHTMLFile('Files/iCBG21FR.html')
    elif option == 'iCW773.html':
        viewHTMLFile('Files/iCW773.html')
        
interact(selectFile, option=widgets.ToggleButtons(options=['iCBG21FR.html', 'iCW773.html'], description='Select File:'))
```
